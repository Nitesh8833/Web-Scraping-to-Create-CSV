```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq

my_url="https://www.flipkart.com/asus-tuf-gaming-core-i5-10th-gen-16-gb-512-gb-ssd-windows-10-home-4-graphics-nvidia-geforce-gtx-1650-ti-144-hz-fx506li-hn279t-laptop/p/itm8bdedba0dbeb6?pid=COMG5QHNHD49JGXG&lid=LSTCOMG5QHNHD49JGXGD2WAFT&marketplace=FLIPKART&store=6bo%2Fb5g&srno=b_1_1&otracker=hp_reco_Laptops%2BFor%2BYou_3_10.dealCard.OMU_cid%3AS_F_NWF_PC8478ab870f0772_af___NONE_ALL%3Bnid%3A6bo_b5g_%3Bet%3APC%3Beid%3APC8478ab870f0772%3Bmp%3AF%3Bct%3Aaf%3B_5&otracker1=hp_reco_WHITELISTED_personalisedRecommendation%2FintentPivoted_Laptops%2BFor%2BYou_DESKTOP_HORIZONTAL_dealCard_cc_3_NA_view-all_5&fm=personalisedRecommendation%2FintentPivoted&iid=62a46179-0337-471a-9e19-50f81911878e.COMG5QHNHD49JGXG.SEARCH&ppt=hp&ppn=homepage&ssid=xmaq6t4jao0000001642923575292"

uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
```

```python
containers = page_soup.findAll("div")
print(len(containers))
```

```python
print(soup.prettify(containers[0]))
```

```python
container = containers[0]
name = container.findAll("div", {"class": "CXW8mj"})
name[0].img["alt"]
```

```python
price = container.findAll("div", {"class": "_25b18c"})
print(price[0].text)
```

```python
ratings = container.findAll("span", {"class": "_2_R_DZ"})
print(ratings[0].text)
```

```python
filename = "products.csv"
f = open(filename, "w")
headers = "Product_Name, Pricing, Ratings \n"
f.write(headers)
```

```python
for container in containers:
    product_name = name[0].img["alt"]
    price_container = container.findAll("div", {"class": "_25b18c"})
    price = price_container[0].text.strip()

    rating_container = container.findAll("span", {"class": "_2_R_DZ"})
    rating = rating_container[0].text
    print("Product_Name:"+ product_name)
    print("Price: " + price)
    print("Ratings:" + rating)
    
```
